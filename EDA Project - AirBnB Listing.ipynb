## Task 5 : Data Analysis
### Univariate Analysis
#### Idendify Outliers In Price
#### A box plot is used to identify outliers in the price column and understand the distribution of listing prices.

```python
df['price']
```

### Outliers

#### The box plot was used to identify outliers in the price column. A few listings had extremely high prices compared to the majority of properties, indicating the presence of price outliers.

```python
data=df[df['price']< 1500]
sns.boxplot(data=data,x='price')
```

### Price Distribution

#### The price distribution chart shows that most Airbnb listings are concentrated in the lower price range, while only a small number of listings have higher prices.

```python
#df['price']
plt.figure(figsize=(8, 5))
sns.histplot(data=data,x='price',bins=100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.dtypes
```

### Availability_365 Distribution

#### This chart illustrates the distribution of listing availability throughout the year. Most properties have limited availability, while fewer listings are available for a large number of days.

```python
plt.figure(figsize=(6, 3))
sns.histplot(data=df,x='availability_365')
plt.title('Availability_365 Distribution ')
plt.ylabel('Frequency')
plt.show()
```

## Heat Map 
### Correlation of one variable with other for anumerical column
#### The heatmap was generated to examine correlations between numerical features. It helps identify variables that have strong positive or negative relationships with each other.

```python
corr=df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds',]].corr()
corr
plt.figure(figsize=(8, 6))
sns.heatmap(data=corr, annot=True)
```
